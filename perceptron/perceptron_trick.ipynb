```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
x, y =make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_classes=2,
                           n_clusters_per_class=1, hypercube=False, random_state=40, class_sep=15)
```

```python
x.shape, y.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x[:,0], x[:,1],c=y, cmap='winter', s=100)
```

```python
def perceptron(x, y):
    x = np.insert(x, 0, 1, axis=1) # adds bias term
    w = np.ones(x.shape[1]) # initializes weights

    lr = 0.1
    for i in range(100):
        r = np.random.randint(0,100)
        y_hat = step(np.dot(x[r], w)) 
        w = w + lr * (y[r] - y_hat) * x[r]

    return w[0], w[1:]
```

```python
def step(z):
    return 1 if z >=0 else 0
```

```python
intercept_, coef_ = perceptron(x, y)
```

```python
print(f"Intercept: {intercept_}, Coefficients: {coef_}")
```

```python
m = -(coef_[0] / coef_[1])
b = -(intercept_ / coef_[1])
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = m * x_input + b
```

```python
plt.Figure(figsize=(10,6))
plt.plot(x_input, y_input, color='red', linewidth=2)
plt.scatter(x[:,0], x[:,1], c=y, cmap='winter', s=100)
plt.ylim(-3, 4)
```
