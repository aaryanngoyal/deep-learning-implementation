```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
x, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_classes=2,
                           n_clusters_per_class=1, hypercube=False, random_state=40, class_sep=15)
x, y
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x[:,0], x[:,1], c=y, cmap='winter', s=100)
```

```python
def perceptron(x, y):
    
    w1=w2=b=1
    lr = 0.1

    for i in range(1000):
        for j in range(x.shape[0]):

            # gradient descent step

            z = w1*x[j][0] + w2*x[j][1] + b
            if z*y[j] < 0:
                w1 += lr * y[j] * x[j][0]
                w2 += lr * y[j] * x[j][1]
                b += lr * y[j]
    
    return w1, w2, b
```

```python
y = np.where(y == 0, -1, 1)
w1, w2, b = perceptron(x, y)
```

```python
m = -(w1 / w2)
c = -(b / w2)
```

```python
x_input = np.linspace(-3, 3, 100)
y = m * x_input + c
```

```python
plt.Figure(figsize=(10,6))
plt.plot(x_input, y, color='red', linewidth=2)
plt.scatter(x[:,0], x[:,1], c=(y+1) // 2, cmap='winter', s=100)
plt.ylim(-3, 4)
```
