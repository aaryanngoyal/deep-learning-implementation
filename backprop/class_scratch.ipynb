```python
import numpy as np 
import pandas as pd
```

```python
df = pd.DataFrame([[8,8,1],[7,9,1],[6,10,0],[5,5,0]], columns=['cgpa', 'profile_score', 'placed'])
df
```

```python
def init_params(layers):

    params={}
    for i in range(1, len(layers)):

        params['W' + str(i)] = np.random.randn(layers[i], layers[i-1]) * 0.1
        params['b' + str(i)] = np.zeros((layers[i], 1))

    return params
```

```python
def sigmoid(z):
    A = 1 / (1 + np.exp(-z))
    return A
```

```python
def forward(A_prev, W, b):
    z = np.dot(W, A_prev) + b
    A = sigmoid(z)
    return A
```

```python
def forward_prop(x, params):
    A = x
    for i in range(1, len(params) // 2 + 1):
        A_prev = A
        Wi = params['W' + str(i)]
        bi = params['b' + str(i)]
        A = forward(A_prev, Wi, bi)

    return A, A_prev
```

```python
def update_params(params, y, y_hat, A1, x):
    params['W2'][0][0] += (0.0001 * (y - y_hat) * A1[0][0])
    params['W2'][0][1] += (0.0001 * (y - y_hat) * A1[1][0])
    params['b2'][0][0] += (0.0001 * (y - y_hat))

    params['W1'][0][0] += (0.0001 * (y - y_hat) * params['W2'][0][0] * A1[0][0] * (1 - A1[0][0]) * x[0][0])
    params['W1'][0][1] += (0.0001 * (y - y_hat) * params['W2'][0][0] * A1[0][0] * (1 - A1[0][0]) * x[1][0])
    params['b1'][0][0] += (0.0001 * (y - y_hat) * params['W2'][0][0] * A1[0][0] * (1 - A1[0][0]))

    params['W1'][1][0] += (0.0001 * (y - y_hat) * params['W2'][0][1] * A1[1][0] * (1 - A1[1][0]) * x[0][0])
    params['W1'][1][1] += (0.0001 * (y - y_hat) * params['W2'][0][1] * A1[1][0] * (1 - A1[1][0]) * x[1][0])
    params['b1'][1][0] += (0.0001 * (y - y_hat) * params['W2'][0][1] * A1[1][0] * (1 - A1[1][0]))
```

```python
parameters = init_params([2,2,1])
epochs = 50

for i in range(epochs):

    Loss = []

    for j in range(df.shape[0]):

        x = df[['cgpa', 'profile_score']].values[j].reshape(2,1)
        y = df[['placed']].values[j][0]

        y_hat, A1 = forward_prop(x, parameters)
        y_hat = y_hat[0][0]

        update_params(parameters, y, y_hat, A1, x)

        Loss.append(-y * np.log(y_hat) - (1 - y) * np.log(1 - y_hat))

    print('Epoch : ', i+1, 'Loss : ', np.array(Loss).mean())
```
