```python
import numpy as np
import pandas as pd 
```

```python
df = pd.DataFrame([[8,8,4],[7,9,5],[6,10,6],[5,12,7]], columns=['cgpa', 'profile_score', 'lpa'])
df
```

```python
def init_params(layers):
    params = {}
    for i in range(1, len(layers)):
        params['W' + str(i)] = np.random.randn(layers[i], layers[i-1]) * 0.01
        params['b' + str(i)] = np.zeros((layers[i], 1))

    return params
```

```python
def forward(W, b, A_prev):
    Z = np.dot(W, A_prev) + b
    return Z
```

```python
def forward_prop(x, params):
    A = x
    for i in range(1, len(params) // 2 + 1):
        A_prev = A

        Wi = params['W' + str(i)]
        bi = params['b' + str(i)]
        A = forward(Wi, bi, A_prev)

    return A, A_prev
```

```python
def update_params(params, y, y_hat, A1, x):
    params['W2'][0][0] += (0.001 * 2 * (y - y_hat) * A1[0][0])
    params['W2'][0][1] += (0.001 * 2 * (y - y_hat) * A1[1][0])
    params['b2'][0][0] += (0.001 * 2 * (y - y_hat))

    params['W1'][0][0] += (0.001 * 2 * (y - y_hat) * params['W2'][0][0] * x[0][0])
    params['W1'][0][1] += (0.001 * 2 * (y - y_hat) * params['W2'][0][0] * x[1][0])
    params['b1'][0][0] += (0.001 * 2 * (y - y_hat) * params['W2'][0][0])

    params['W1'][1][0] += (0.001 * 2 * (y - y_hat) * params['W2'][0][1] * x[0][0])
    params['W1'][1][1] += (0.001 * 2 * (y - y_hat) * params['W2'][0][1] * x[1][0])
    params['b1'][1][0] += (0.001 * 2 * (y - y_hat) * params['W2'][0][1])
```

```python
parameters = init_params([2,2,1])
epochs = 50

for i in range(epochs):

    Loss = []

    for j in range(df.shape[0]):

        x = df[['cgpa', 'profile_score']].values[j].reshape(2,1)
        y = df[['lpa']].values[j][0]

        y_hat, A1 = forward_prop(x, parameters)
        y_hat = y_hat[0][0]

        update_params(parameters, y, y_hat, A1, x)

        Loss.append((y - y_hat) ** 2)

    print('Epoch : ', i+1, 'Loss : ', np.array(Loss).mean())
```
